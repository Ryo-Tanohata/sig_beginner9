```python
#signate_begginer9.zipをGoogle Drive からひっぱってくるやつ
from googleapiclient.discovery import build
import io, os
from googleapiclient.http import MediaIoBaseDownload
from google.colab import auth
 
auth.authenticate_user()
 
drive_service = build('drive', 'v3')
results = drive_service.files().list(
        q="name = 'signate_begginer9.zip'", fields="files(id)").execute()
signate_api_key = results.get('files', [])
 
filename = "/content/input/signate_begginer9.zip"
os.makedirs(os.path.dirname(filename), exist_ok=True)
 
request = drive_service.files().get_media(fileId=signate_api_key[0]['id'])
fh = io.FileIO(filename, 'wb')
downloader = MediaIoBaseDownload(fh, request)
done = False
while done is False:
    status, done = downloader.next_chunk()
    print("Download %d%%." % int(status.progress() * 100))
os.chmod(filename, 600)
```

```python
import os
path="./input"
zip_files=os.listdir(path)
zip_files
```

```python
import zipfile
for i in zip_files:
    if ".zip" in i:
        with zipfile.ZipFile("input/"+i) as existing_zip:
            print(f'open {i}')
            existing_zip.extractall('input')
            print("extract")
    else:
        continue
```

```python
import os
datapath=[]
for dirname, dirpaths, filenames in os.walk('input'):
    for filename in filenames:
        datapath.append(dirname+'/'+filename)
datapath
```

```python
import pandas as pd
databox={}
for i in datapath:
    if '.tsv' in i:
        data={i.split('/')[1].split('.')[0]:pd.read_csv(i, sep='\t', index_col=0)}
    elif  ('train' in i) or ('test' in i):
        data={i.split('/')[1].split('.')[0]:pd.read_csv(i)}
    elif '.csv' in i:
        data={i.split('/')[1].split('.')[0]:pd.read_csv(i, header=None)}
    databox.update(data)
databox.keys()
```

```python
train=databox['train']
test=databox['test']
sub=databox['sample_submission']
```

```python
train.head()
```

```python
y_col=list(train.columns)
for i in list(test.columns):
    y_col.remove(i)
y_col
```

```python
features=list(train.columns)
features.remove(y_col[0])
features
```

```python
X=train[features]
y=train[y_col]
```

```python
import sys
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from sklearn.utils.testing import all_estimators
##############################
# 全モデルでK-分割交差検証を行う
# K-fold cross-validation with all estimators.
##############################
# x_train = df.drop(columns='Survived').values
# y_train = df[['Survived']].values
# y_train = numpy.ravel(y_train)

kf = KFold(n_splits=3, shuffle=True, random_state=1)

# with open('./all_estimators_classifier.txt', 'w', encoding="utf-8") as f:
#     f.write('name\taccuracy\n')
# writer = open('./all_estimators_classifier.txt', 'w', encoding="utf-8")
# writer.write('name\taccuracy\n')
all_models=pd.DataFrame(columns=["name","accuracy"])
for (name,Estimator) in all_estimators(type_filter="classifier"):

    try:
        model = Estimator()
        if 'score' not in dir(model):
            continue
        else:
            scores = cross_validate(model, X, y, cv=kf, scoring=['accuracy'])
            accuracy = scores['test_accuracy'].mean()
            print(name, accuracy)
            all_models.append({'name':str(name),'accuracy':accuracy},ignore_index=True)
            # writer.write(name + "\t" + str(accuracy) + '\n')
    except:
        print(sys.exc_info())
        print(name)
        pass

# writer.close()
```
